```python
# The second is a grouped bar chart for Midfielders. Positions (LM, RM, CM)

Mid_Position=Nationality[(Nationality["BestPosition"] == "LM") | (Nationality["BestPosition"] == "RM") | (Nationality["BestPosition"] == "CM")].copy()
Mid_Position["Count"]=1
Mid_Position=pd.pivot_table(Mid_Position,values="Count",index=['Nationality'], columns = 'BestPosition',aggfunc="sum")
Mid_Position.plot.bar()
```

```python
#The third plot is a grouped bar chart for Defenders. Positions (LB, RB, CB)

Defender_Position=Nationality[(Nationality["BestPosition"] == "LB") | (Nationality["BestPosition"] == "RB") | (Nationality["BestPosition"] == "CB")].copy()
Defender_Position["Count"]=1
Defender_Position=pd.pivot_table(Defender_Position,values="Count",index=['Nationality'], columns = 'BestPosition',aggfunc="sum")
Defender_Position.plot.bar()
```
